```python
from google.colab import files
uploaded = files.upload()
import cv2
import matplotlib.pyplot as plt
from io import BytesIO
from PIL import Image
import numpy as np
```

```python
#This is a key generation function for logistic map(chaotic map)
# It will gernerate 256*256 pixel image into  a 2-D matrix number form
def keygen(x,r,size):
  key=[]
  for i in range (size):
    x=r*x*(1-x)
    key.append(int(x*(pow(10,16))%256))
  return key
```

```python
img = cv2.imread("cameraman.bmp")
img_cvt=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
plt.show()



print(img_cvt.shape)
```

```python
img = cv2.imread("cameraman.bmp")
img_cvt=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
plt.show()
height=img_cvt.shape[0]
width=img_cvt.shape[1]
#height and width of image
print(height)
print(width)

#Generate  key for every  point of (256*256)image 

key=keygen(0.01,3.95,height*width)
print("Encryption key for every bit of image\n")
print(key)


#Encryption key is generated and 

# print(img_cvt.shape)
```

```python
img = cv2.imread("cameraman.bmp")
img_cvt=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
plt.show()
height=img_cvt.shape[0]
width=img_cvt.shape[1]

print("Origional image Histogram:")
img = cv2.imread("cameraman.bmp")
img_cvt=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
histr =cv2.calcHist([img],[0],None,[256],[0,256]) 
plt.plot(histr)
plt.show()

#height and width of image
#print(height)
#print(width)

#Generate  key for every  point of (256*256)image 

#key=keygen(0.01,3.95,height*width)
#print("Encryption key for every bit of image\n")
#print(key)


#Encryption key is generated and 

# print(img_cvt.shape)

#Applying uisng Xor encryption algorithm
z=0
encrypt_image=np.zeros(shape=[height,width,3],dtype=np.uint8)
for i in range(height):
  for j in range(width):
    encrypt_image[i,j]=img_cvt[i,j]^key[z]
    z+=1



print("Encrypted image uisng chaotic key")
plt.imshow(encrypt_image)
plt.show()
plt.imsave('encrypt1.bmp',encrypt_image)

print("Encrypted  image Histogram:")
img = cv2.imread("encrypt1.bmp")
img_cvt=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
histr =cv2.calcHist([img],[0],None,[256],[0,256]) 
plt.plot(histr)
plt.show()

print("Decrypted image using same key\n")



z=0
decrypt_image=np.zeros(shape=[height,width,3],dtype=np.uint8)
for i in range(height):
  for j in range(width):
    decrypt_image[i,j]=encrypt_image[i,j]^key[z]
    z+=1


plt.imshow(decrypt_image)
plt.show()
plt.imsave('decrypt1.bmp',decrypt_image)


```
