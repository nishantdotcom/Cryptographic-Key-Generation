```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python

r=3

N=500
x=np.zeros(N)
x[0]= 0.5



for n in range(N-1):
  x[n+1]=r*x[n]*(1-x[n])


plt.plot(x,'-')
plt.show()


```

```python


rs=np.linspace(1,4,1000)

N=500
x=np.zeros(N)
x[0]= 0.5
endcap=np.arange(round(N*.9),N)


for ri in range(len(rs)):
  for n in range(N-1):
    x[n+1]=rs[ri]*x[n]*(1-x[n])
  
  u=np.unique(x[endcap])
  r=rs[ri]*np.ones(len(u))
  plt.plot(r,u,'k.',markersize=1)



plt.show()
```
